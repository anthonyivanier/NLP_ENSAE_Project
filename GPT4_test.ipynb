```python
# Compute the accuracy of the predictions of gpt4 vs sex column of df_transcriptions
def accuracy(data_pred_gpt, df):
    n = len(data_pred_gpt)
    correct = 0
    for i in range(n):
        if data_pred_gpt[i]['sexe_prédit'] == df['sex_en_anglais'][i]:
            correct += 1
    return correct/n

accuracy(data_pred_gpt, df)
```
